# Aims

As a software application or device, IDS observes network for detection malicious violations and activity. In modern society, where information is property and goods, it is an important means of keeping our information and most of companies and research institutions are implementing IDS on their intranet or network. Among various malicious activities, malware has the most variety of forms, ranging from classic computer viruses to Internet worms and bot networks. Currently, most of IDS approaches for malware detection is divided into signature approach and rule-based approach. Signature approach IDS works similarly to virus scanners by scanning the signature database for known identities or signatures for each specific malware intrusion. Malwares monitored by signature IDS are matched against malware signature databases to detect intrusions. Due to this attribute, signature approach cannot detect a new unknown malware attack because it must manually modify the signature database for each new type of intrusion found. On the other hand, rule-based approach follows 'if-then' rules which on the basis of cybersecurity expert's knowledge that is earned from numerous vuluerabilities and intrusion events. In rule-based approach, malware is typically monitored and then transformed into facts and rules that are later employed by inference algorithm to reach conclusions. However, sometimes the converted rules is vague and it causes high rate of false/negative and false/positive. As a result, it can put heavy burden tasks on the system administer. Also, like signature approach, it tend to show insufficient respond to the advert of variety of malware, which caused by proliferation technology.

Currnetly, generated by application metamorphosis and polymorphism technology, varient of malware looks like totally different from existing malware, but it has same malicious affect on the system or netowrk. Signature approach and rule-based approach are vulnerable to this kind of malware. Therefore, this project proposes new approach for malware detection to deal with varient of malware more efficiently. The main goal of this project is implementing visualization approach with machine learning on Intrusion Detection System(IDS) for malware detection. The aim can be divided into four objectives: (1) generate fast and efficient transform methods for use in real-time, from binary exectable file to gray scale image; (2) make balanced training dataset; (3) implement diverse models and evaluate the models for malware detetion; and (4) validate the approach's performance on malware detection and deploy without conflict on existing IDS.

# **Background**
As the activities, performed through face-to-face contact in the past, are happening in the cyber environment due to the spread of the Internet such as shopping, communication with friends and payment, cyber-crime have became one of most serious problem in modern society(Chen et al. 2018). Especially, to companies and research institutions, which deal with high level of intellectual property, cyber-threats are more fatal. According to Cybersecurity Ventures' 2019 ACR, the damage caused by cybercrime is anticipated to arouse $ 6 trillion corporate and organizational costs annually until 2021(Casey 2019). Most organizations that recognize these threats are using Intrusion Detection Systems(IDS) to monitor their networks and detect and prevent malicious cyber threats. Commercial IDS used in the current industry is largely divided into signature approach and rule-based approach(Dali et al. 2015). Signature approach IDS works similarly to virus scanners by scanning the signature database for known identities or signatures for each specific malware intrusion. Malwares monitored by signature IDS are matched against malware signature databases to detect intrusions. Due to this attribute, signature approach cannot detect a new unknown malware attack because it must manually modify the signature database for each new type of intrusion found. On the other hand, rule-based approach follows 'if-then' rules which on the basis of cybersecurity expert's knowledge that is earned from numerous vuluerabilities and intrusion events. In rule-based approach, malware is typically monitored and then transformed into facts and rules that are later employed by inference algorithm to reach conclusions. However, sometimes the converted rules is vague and it causes high rate of false/negative and false/positive. As a result, it can put heavy burden tasks on the system administer. Also, because the expert knowledge that underlies the rule-based model builds on the experience of working with previous malware, rule-based model can't respond correctly to newer forms of malware. Based on these facts, both approaches are vulnerable to new malware.
Attackers who find this vulnerability use metamorphosis and polymorphism technology to generate malware that proliferates, leading to an increase in variants of malware(Figure_1).

In 2016, Lakshmanan came up with a new approach to malware(Nataraj et al. 2011). It is the visualization approach. By using the characteristic of malware executable file, which can be represented as a binary value(ones and zeros), malware executable file is transformed as grey scale image. Through the images obtained from transforming, there is a significant similarity between variants of malware with the same malicious effect and a distinct difference between malwares with different malicious effects(Figure_2)

```
Figure_2
```

In other words, it is more efficient and flexible to classify malware on transformed gray scale image(visualization approach) analysis than on raw binary files(signature approach and rule-based approach) analysis. Furthermore, they found unexpected benefit from visualization approach. In some case, for obfuscation on raw malware, the malware is packed, which change the original form of malware file. It causes confusion on the analysis malware in signature approach and rule-based approach. On the other hand, visualization approach relies mainly on image pattern information preserved by the weak encryption scheme used by the polymorphic engine, thus, can successfully handle encrypted section.

# **Research Project**
Among the cyber threats to Internet environment these days, malicious software, called malware, can cause most serious effects on the computer system. Malicious methods, varying from typical form of viruses to Internet worms and bot networks, aim at devices connected to the network. Most of this threat comes from the criminal industry. To achieve outlawed goals, such as collecting confidential data and spreading spam mails, the criminal industry comprehensive organizes networked hosts(Franklin et al., 2007; Holz et al., 2009). Regrettably, as the amount and variety of malware increased, preexistence approaches, such as signature approach and rule-based approach, became inadequate, as a result, leaving millions of hosts on the cyber space infected with malicious software. This can disrupt the continuity of your business and can have a negative impact on other hosts in your network. For example, in January 2003, the Slammer worm, which is kind of malware, induced a geographically rotating Internet blackout in the Australia, New Zealand, Korea, and United States. The reckless spread of the worm led to a 25% increase in network traffic, causing severe issues for Bank of America's bank operations. Network worm epidemics such as Blaster, MSBlast, Mydom, and Sasser have caused enormous damage to airlines that have to cancel flights and banks that need to be temporarily shut down. Therefore, to modern companies and governments, the signature approach and rule-based approach, currently used for malware detection and classification in commercial IDS, need to be replaced with a more flexible and efficient way to handle large amounts of data, or to address the problems of existing methods.

To solve the problem cause by proliferated variant of malware, this project proposes visualization approach. However, unlike prior research, in the project, we are going to implement Convolutional Neural Network (CNN) models instead of basic machine learning classifier methods, such as K-Nearest Neighbor(KNN), Support Vector Machines(SVM), Decision Trees and Random forest. CNN is a type of Deep Learning(DL) algorithm that consists of one or several convolutional layers, a pooling layer, and fully connected layers, except input layer and output layer(Figure_3).

```
Figure_3
```

CNN is based on the process of inductively setting inference structure or learning parameters (weights) for optimal inference based on given training data(Keren & Schuller 2016). The convolutional layer is the core element in the CNN structure, and is a layer for extracting features of data. Different concepts of CNN model are classified according to the structure of convolutional layer. In the pooling layer, this layer abstracts the feature map, extracted from convolutional layer. In this layer, dimension reduction through sub-sampling is executed, and it makes possible to compute more quickly. Fully connected layers apply in the top layers and classify the features conveyed in the lower layers. There are 3 reasons why we choose the CNN to visualization approach for malware analysis, instead of basic machine learning classifier methods. (1) CNN model needs much lower pre-processing step. This characteristic allows IDS to analyze large amount of data quickly in real-time; (2) CNN shows excellent performance in the field of image recognition and classification. According to result of ImageNet Large Scale Visual Recognition Challenge(ILSVRC) from 2010, CNN outperforms hand-crafted descriptors and shallow networks since 2012(Nguyen et al. 2017); (3) CNN has various concept algorithms such as AlexNet, VGG, GoogleNet and ResNet, which can be applied according to the characteristics of the data to find a system that is suitable on the purpose. Also, more efficient CNN algorithms are still being actively researched. Therefore, it is more likely to develop than now.
